```python
i='ApplicantIncome'
q1=np.percentile(df[i],25)
q3=np.percentile(df[i],75)

iqr=q3-q1

c1=q1-1.5*(iqr)
c2=q1+1.5*(iqr)
```

```python
w=df[(df[i]>c2)|(df[i]<c1)]
```

```python
w=w.index
```

```python
w
```

```python
df=df.drop(labels=w,axis=0)
```
